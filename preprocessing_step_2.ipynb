## This code does the following:
1. Apply YOLOv5 on cat and dog images,

2. Detect the relevant object (cat or dog), eliminating those that fail to reach the 50% threshhold of quality. 

3. Crop the bounding box,

4. Resize it to 224√ó224,

5. Save the cleaned image in an output folder (cropped_cat_images, cropped_dog_images).

```python
import torch
import os
import cv2

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
model.conf = 0.5  # Confidence threshold (skip low-quality detections)

# Input and output paths
input_folders = {
    'cat': 'cat_images',
    'dog': 'dog_images'
}

output_base = 'cropped_images'
os.makedirs(output_base, exist_ok=True)

LABEL_MAP = {
    'cat': 15,
    'dog': 16
}

for label, folder in input_folders.items():
    input_path = folder
    output_path = os.path.join(output_base, f'{label}_images')
    os.makedirs(output_path, exist_ok=True)

    print(f"\nProcessing {label.upper()} images...")

    for filename in os.listdir(input_path):
        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue

        file_path = os.path.join(input_path, filename)
        image = cv2.imread(file_path)
        if image is None:
            continue

        # Run YOLOv5 detection
        results = model(image)
        detections = results.xyxy[0]

        found = False
        for *box, conf, cls in detections:
            if int(cls) == LABEL_MAP[label]:
                x1, y1, x2, y2 = map(int, box)
                cropped = image[y1:y2, x1:x2]
                resized = cv2.resize(cropped, (224, 224))
                save_path = os.path.join(output_path, filename)
                cv2.imwrite(save_path, resized)
                found = True
                break

        if not found:
            print(f"‚ö†Ô∏è  No clear {label} detected in {filename}")

```
